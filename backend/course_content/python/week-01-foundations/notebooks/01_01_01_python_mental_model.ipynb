## Part 1 — Everything is an object

Every value in Python — every integer, string, function, class, `None` — lives as an **object** in memory. An object has exactly three properties:

| Property | What it is | How to inspect |
|---|---|---|
| **Identity** | Unique ID — its address in memory | `id(obj)` |
| **Type** | What kind of object it is | `type(obj)` |
| **Value** | The data it holds | Depends on type |

### The one thing that changes everything

> **A variable is not a box that holds a value.**
> **A variable is a name — a post-it note — attached to an object.**

### Visualised

```
WRONG mental model (JavaScript / Java box model):

    ┌─────────────────────┐
    │  x  │   42          │   x is a box. 42 lives inside it.
    └─────────────────────┘


CORRECT mental model (Python label model):

                    ┌──────────────────┐
    x ─────────────►│  int: 42  id=0xA │   x is a label. 42 is a separate object.
                    └──────────────────┘
```

When you write `y = x`:
```
                    ┌──────────────────┐
    x ─────────────►│  int: 42  id=0xA │   Both labels point to the same object.
    y ─────────────►│                  │   No copy was made.
                    └──────────────────┘
```

When you write `x = 99`:
```
    x ─────────────►┌──────────────────┐
                    │  int: 99  id=0xB │   x label moves. y label stays.
                    └──────────────────┘

    y ─────────────►┌──────────────────┐
                    │  int: 42  id=0xA │   Same object. Untouched.
                    └──────────────────┘
```

> 🗂️ **Sticky phrase — Part 1:** "`=` moves a label. It never copies an object."

> ⚠️ **Common mistake (Java/JS engineers):** Assuming `y = x` gives `y` its own independent copy. It does not. The copy only happens if you ask for one explicitly. You will see exactly why this matters when we hit Part 2.


````python
# Run this — verify it matches the diagrams above

x = 42
y = x

print("After x = 42, y = x:")
print(f"  id(x) = {id(x)}")
print(f"  id(y) = {id(y)}")
print(f"  x is y: {x is y}  <-- same object, two labels")

x = 99

print("\nAfter x = 99:")
print(f"  x = {x},  id(x) = {id(x)}")
print(f"  y = {y},  id(y) = {id(y)}  <-- y label never moved")
print(f"  x is y: {x is y}  <-- different objects now")
````

## Part 2 — Mutable vs Immutable

| Immutable (cannot change in place) | Mutable (can change in place) |
|---|---|
| `int`, `float`, `bool`, `complex` | `list` |
| `str` | `dict` |
| `tuple`, `frozenset` | `set` |
| `bytes` | Most class instances |

**The critical difference:** when you pass a **mutable** object to a function, the function can modify the original. Pass an **immutable** object and it cannot — any "modification" creates a new object.

### The aliasing trap — visualised

This is the shape of Priya's bug. Study the diagram before reading the code.

```
STEP 1: a = ["contract_A.pdf"]

    a ─────────────►┌─────────────────────────┐
                    │  list: ["contract_A.pdf"] │  id=0xC
                    └─────────────────────────┘

STEP 2: b = a     (NOT a copy — just a second label)

    a ─────────────►┌─────────────────────────┐
    b ─────────────►│  list: ["contract_A.pdf"] │  id=0xC
                    └─────────────────────────┘

STEP 3: b.append("contract_B.pdf")
         (.append changes the object IN PLACE — no new object created)

    a ─────────────►┌───────────────────────────────────────┐
    b ─────────────►│  list: ["contract_A.pdf","contract_B.pdf"] │  id=0xC
                    └───────────────────────────────────────┘

Both labels see the change. You didn't copy anything. You just added a label.
```

> 🗂️ **Sticky phrase — Part 2:** "Mutation changes the object itself. Every label pointing to it sees the change."

> ⚠️ **Common mistake:** Thinking `b = a` is "safe" because you're working through `b`. It is not. Both labels point to the same object. There is no isolation.


````python
# The aliasing trap

a = ["contract_A.pdf"]
b = a                         # second label — same object

b.append("contract_B.pdf")   # mutates the object in place

print(f"a = {a}")   # Expected: ["contract_A.pdf", "contract_B.pdf"]
print(f"b = {b}")   # Expected: ["contract_A.pdf", "contract_B.pdf"]
print(f"a is b: {a is b}")  # Expected: True — same object, two labels
````

## Part 3 — Shallow vs Deep copy

A **shallow copy** creates a new outer container — but the objects inside it are still shared.
A **deep copy** recursively creates new objects at every level.

### When does this matter?
Only when you have **nested mutable objects** — a list of lists, a dict of lists, etc.
For flat structures (list of strings), shallow copy is completely safe.

### Visualised

```
SHALLOW COPY (e.g., batch_B = batch_A[:]):

    batch_A ──►┌─────────────────────────────────────┐  id=0xD
               │  [0] ──►┌────────────────┐ id=0xE  │
               │         │ "c1", "c2"     │ SHARED  │
               │         └────────────────┘         │
               │  [1] ──►┌────────────────┐ id=0xF  │
               │         │ "c3"           │ SHARED  │
               └─────────└────────────────┘─────────┘

    batch_B ──►┌─────────────────────────────────────┐  id=0xG  (NEW outer)
               │  [0] ──►┌────────────────┐ id=0xE  │  SAME inner objects
               │         │ "c1", "c2"     │         │
               │         └────────────────┘         │
               │  [1] ──►┌────────────────┐ id=0xF  │
               │         │ "c3"           │         │
               └─────────└────────────────┘─────────┘

Mutating batch_B[0] mutates id=0xE — which batch_A[0] also points to.


DEEP COPY (e.g., batch_C = copy.deepcopy(batch_A)):

    batch_C ──►┌─────────────────────────────────────┐  id=0xH  (NEW outer)
               │  [0] ──►┌────────────────┐ id=0xI  │  NEW inner objects
               │         │ "c1", "c2"     │         │
               │         └────────────────┘         │
               │  [1] ──►┌────────────────┐ id=0xJ  │
               │         │ "c3"           │         │
               └─────────└────────────────┘─────────┘

Nothing shared. Changes to batch_C never affect batch_A.
```

> 🗂️ **Sticky phrase — Part 3:** "Shallow copies the wrapper. Deep copies the contents. Flat data? Shallow is fine."

> ⚠️ **Common mistake:** Reaching for `deepcopy` as the default safe option. It's slow and usually signals a design problem. In 15 years I've seen `deepcopy` used correctly in production GenAI code maybe twice. The right fix is almost always "return a new object instead of mutating a shared one."


````python
import copy

# Nested list — list of contract batches
batch_A = [["contract_1.pdf", "contract_2.pdf"], ["contract_3.pdf"]]
batch_B = batch_A[:]   # shallow copy — new outer, shared inner lists

# Safe: appending to the outer list
batch_B.append(["contract_4.pdf"])
print(f"batch_A len: {len(batch_A)}  (Expected: 2 — outer append is safe)")
print(f"batch_B len: {len(batch_B)}  (Expected: 3)")

# Danger: mutating an inner list — both see the change
batch_B[0].append("contract_0.pdf")
print(f"\nbatch_A[0]: {batch_A[0]}  (Expected: 3 items — CHANGED via shared ref)")
print(f"batch_B[0]: {batch_B[0]}  (Expected: 3 items)")
print(f"Same inner? {batch_A[0] is batch_B[0]}  (Expected: True)")
````

````python
import copy

batch_A = [["contract_1.pdf", "contract_2.pdf"], ["contract_3.pdf"]]
batch_C = copy.deepcopy(batch_A)   # fully independent at every level

batch_C[0].append("contract_0.pdf")

print(f"batch_A[0]: {batch_A[0]}  (Expected: 2 items — untouched)")
print(f"batch_C[0]: {batch_C[0]}  (Expected: 3 items)")
print(f"Same inner? {batch_A[0] is batch_C[0]}  (Expected: False)")
````

## Part 4 — `is` vs `==`

| Operator | Checks | Correct use |
|---|---|---|
| `==` | **Value equality** — same value? | Comparing any values |
| `is` | **Identity** — literally the same object in memory? | `None`, `True`, `False` only |

The rule is absolute:
- `is` for `None`, `True`, `False` — Python guarantees these are singletons
- `==` for everything else

**Why `is` appears to work for small integers (and why that's dangerous):**

CPython **interns** (caches) integers from −5 to 256 and short strings as an optimisation. `is` appears to return `True` for these values. But this is a CPython implementation detail — not a Python language guarantee. It fails for values outside that range. Code that relies on it is a latent bug waiting to trigger in production.

```
After `result = None`:

    result ──────────────────►┌──────────────────┐
                              │  NoneType: None  │  id=0x5560  (singleton)
                              └──────────────────┘

    None  ────────────────────► (same object — always)

result is None   →  True   ✓  checks identity of the singleton
result == None   →  True   but triggers a linter warning — use `is`
```

> 🗂️ **Sticky phrase — Part 4:** "`is` checks if two labels point to the same object. Use it only for `None`."

> ⚠️ **Common mistake (from Java):** Using `==` to compare object identity. In Python, `==` checks value — not identity. The reverse of Java's default. If you want identity, use `is`. If you want value equality, use `==`. They do not overlap except for singletons.

## Part 5 — The mutable default argument bug

> You now have the mental model. This is Priya's bug. You will understand it completely.

**The rule:**

Default argument values are evaluated **once**, at function **definition** time — not at call time.

This means a mutable default (a list `[]`, a dict `{}`) is the **same object** on every call that doesn't provide that argument explicitly.

Here is what Python does when it executes `def process_contract(filename, results=[])`:

```
Definition time (happens once):

    process_contract.__defaults__ = ( list_object,  )
                                          │
                                          └──► ┌────────────────┐
                                               │  list: []      │  id=0xX
                                               └────────────────┘

Call 1:  results ──────────────────────────────► ┌──────────────────────┐
                                                  │  [{"file":"NDA"}]   │  id=0xX
                                                  └──────────────────────┘

Call 2:  results ──────────────────────────────► ┌──────────────────────────────────┐
                                                  │  [{"file":"NDA"},               │  id=0xX
                                                  │   {"file":"MSA"}]               │
                                                  └──────────────────────────────────┘

Same object. Growing forever.
```

> 🗂️ **Sticky phrase — Part 5:** "The list `[]` as a default is shared. `None` is not. Use `None`."
>
> Remember Part 1's sticky phrase? "`=` moves a label." This is that principle applied to default arguments — the label `results` gets moved to the same pre-existing list object on every call. You've seen this exact mechanism twice already.

> ⚠️ **Common mistake:** Thinking this only matters if calls are concurrent. It doesn't. Sequential calls to the same function accumulate state in the shared default. The bug manifests on call 2 regardless of concurrency.

The `results=[]` default is one list object that lives on `process_contract.__defaults__`. Every call that omits `results` appends to that same object. Apply the diagram above.

````python
# THE BUG — run after you have written your predictions above

def process_contract(filename: str, results: list = []) -> list:
    # [] is created ONCE at definition time. Shared across all calls.
    results.append({"filename": filename, "status": "processed"})
    return results


call_1 = process_contract("contract_NDA.pdf")
print(f"Call 1: {call_1}")
# Expected: [{"filename": "contract_NDA.pdf", "status": "processed"}]

call_2 = process_contract("contract_MSA.pdf")
print(f"Call 2: {call_2}")
# Expected if bug: both items -- accumulation

call_3 = process_contract("contract_SOW.pdf")
print(f"Call 3: {call_3}")
# Expected if bug: all three items

# Prove it is the __defaults__ object
print(f"\nDefault object id:       {id(process_contract.__defaults__[0])}")
print(f"call_1 id:               {id(call_1)}")
print(f"All same object:         {process_contract.__defaults__[0] is call_1}")
````


### The three canonical fixes

**Fix 1 — None sentinel** *(use this by default)*


````python
from __future__ import annotations
from typing import Optional


def process_contract_v1(
    filename: str,
    results: Optional[list] = None,   # None is immutable — safe default
) -> list:
    if results is None:
        results = []    # new list created fresh on each call
    results.append({"filename": filename, "status": "processed"})
    return results


# Verify independence
c1 = process_contract_v1("contract_NDA.pdf")
c2 = process_contract_v1("contract_MSA.pdf")
c3 = process_contract_v1("contract_SOW.pdf")

print(f"Call 1: {c1}  (Expected: 1 item)")
print(f"Call 2: {c2}  (Expected: 1 item — independent)")
print(f"Call 3: {c3}  (Expected: 1 item — independent)")
assert len(c1) == len(c2) == len(c3) == 1, "Fix failed — still sharing state"
print("Independence verified.")
````


**Fix 2 — Functional: return one result, let the caller accumulate** *(cleanest design)*

## Part 6 — Truthy and Falsy

Python evaluates objects in boolean context according to one consistent rule.

**Falsy** (evaluates to `False`): `None`, `False`, `0`, `0.0`, `0j`, `""`, `[]`, `()`, `set()`, `{}`

**Everything else is truthy.**

Critical differences from JavaScript:
- `[]` is **falsy** in Python — **truthy** in JavaScript
- `"0"` is **truthy** in Python (non-empty string) — **falsy** in JavaScript
- `{}` (empty dict) is **falsy** in Python — **truthy** in JavaScript (`{}` is a truthy object in JS)

## Part 7 — Scoping: LEGB

When Python encounters a name, it searches four scopes in order and stops at the first match:

```
L  Local       — inside the current function
E  Enclosing   — in any outer function (for nested functions only)
G  Global      — at the module level
B  Built-in    — Python's built-in names: len, print, range, ...
```

### The critical difference from JavaScript — no block scope

```
JavaScript:                          Python:
──────────────────────────────────   ────────────────────────────────────
function f() {                       def f():
    if (true) {                          if True:
        let x = 10;  ← x lives here         x = 10    ← x lives in f()
    }
    console.log(x);  // ReferenceError   print(x)     # works — x visible
}
```

**In Python, `for` loops, `with` blocks, `if` statements, and `try` blocks do NOT create new scopes.** Variables defined inside them belong to the enclosing function (or module).

### Visualised — LEGB lookup

```
def make_request(endpoint: str) -> str:         ← L scope starts here
    timeout = 30

    def build_url() -> str:                     ← new L scope
        # Looking up "BASE_URL":
        # L: not in build_url ──► miss
        # E: not in make_request ──► miss
        # G: found! BASE_URL at module level ──► use it

        # Looking up "timeout":
        # L: not in build_url ──► miss
        # E: found! timeout in make_request ──► use it

        return f"{BASE_URL}{endpoint}?timeout={timeout}"
    ...
```

> 🗂️ **Sticky phrase — Part 7:** "LEGB. Stop at the first hit. No block scope — loop variables leak."

> ⚠️ **Common mistake (JavaScript engineers):** Expecting a `for` loop variable to be undefined after the loop. In Python, it persists in the enclosing scope. This is almost never a bug in practice, but it surprises JS engineers the first time they see it.


````python
BASE_URL = "https://api.anthropic.com/v1"   # G — global

def make_request(endpoint: str) -> str:
    timeout = 30                              # L — local to make_request

    def build_url() -> str:
        # E: endpoint and timeout from make_request
        # G: BASE_URL from module level
        return f"{BASE_URL}{endpoint}?timeout={timeout}"

    return build_url()


result = make_request("/messages")
print(result)
# Expected: https://api.anthropic.com/v1/messages?timeout=30
````

## Part 8 — Python's execution model and the GIL

You need exactly enough of this to make the right architectural decision in Week 3, when you build an async LLM client. No more, no less.

### The three facts

1. Python source is compiled to **bytecode**. CPython (the standard Python interpreter) executes the bytecode.
2. CPython has a **Global Interpreter Lock (GIL)** — a mutex that allows only **one thread** to execute Python bytecode at a time.
3. The GIL is **released** when a thread is waiting for I/O — a network response, a file read, a database query.

### The consequence for your work

| Workload type | Example | Threads help? | Right tool |
|---|---|---|---|
| CPU-bound | Parsing 10k PDFs locally | No — GIL blocks | `multiprocessing` |
| I/O-bound | 50 Anthropic API calls | Yes — GIL released | `asyncio` |
| I/O-bound | Reading from S3, Qdrant | Yes — GIL released | `asyncio` |

**Virtually all GenAI work is I/O-bound.** You are waiting for the LLM API, the vector database, the S3 bucket — not computing locally.

### The bridge to Week 3

In **Module 3.2** you will build an async LLM client that fires 50 contract extraction calls concurrently and finishes in roughly the same time as a single serial call. That 50× throughput is possible because of point 3 above: every `await` on a network call releases the GIL, allowing the event loop to start the next request immediately.

The object model you built this morning — labels, objects, mutation — is what lets you reason about what is shared between coroutines and what isn't. The same mental model. A new application.

> 🗂️ **Sticky phrase — Part 8:** "CPU work: GIL blocks threads. I/O work: GIL releases. LLM calls are I/O. asyncio wins."


````python
import threading
import time

COUNT = 3_000_000

def cpu_work(n: int) -> None:
    """Pure CPU — just counting. No I/O. GIL never releases."""
    total = 0
    for _ in range(n):
        total += 1

# Serial baseline
start = time.perf_counter()
cpu_work(COUNT)
cpu_work(COUNT)
serial = time.perf_counter() - start

# Threaded — same total work, but GIL means only one thread runs at a time
start = time.perf_counter()
threads = [threading.Thread(target=cpu_work, args=(COUNT,)) for _ in range(2)]
for t in threads: t.start()
for t in threads: t.join()
threaded = time.perf_counter() - start

print(f"Serial:   {serial:.2f}s")
print(f"Threaded: {threaded:.2f}s")
print(f"Speedup:  {serial/threaded:.2f}x  (expect ~1.0 -- GIL prevents CPU parallelism)")
print()
print("In Module 3.2: same measurement with async LLM calls -> 20-50x speedup.")
print("The only difference: LLM calls release the GIL. CPU work does not.")
````

````python
# Why the GIL exists -- bytecode shows you
# The GIL protects CPython's reference counts between bytecode instructions.
# Every object tracks how many labels point to it (the reference count).
# The GIL prevents two threads from modifying that count simultaneously.

import dis

def get_firm_name(contract: dict) -> str:
    return contract["firm"].upper()

print("Each line of bytecode is one step CPython executes.")
print("The GIL ensures no two threads execute the same step simultaneously.")
print("Between any two of these instructions, a reference count could be at risk.")
print()
dis.dis(get_firm_name)
````
